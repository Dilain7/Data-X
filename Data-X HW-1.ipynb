```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib as plt 
import sklearn
from sklearn.model_selection import train_test_split
```

```python
# Reading all the relevant Datasets

customers = pd.read_csv('olist_customers_dataset.csv')
order_items = pd.read_csv('olist_order_items_dataset.csv')
products = pd.read_csv('olist_products_dataset.csv')
orders = pd.read_csv('olist_orders_dataset.csv')
```

# Data Check/update step 1

```python
# Merging datasets to make a centralized one with all the relevant data

dataframe = pd.merge(products, order_items, on = 'product_id')
```

# Data Check / update step 2

```python
dataframe = dataframe.dropna()
dataframe = dataframe.drop(columns = ['order_id', 'order_item_id', 'seller_id', 'shipping_limit_date'])
```

# Data Check / update step 3

```python
dataframe = dataframe.drop(columns = ['product_category_name', 'product_id', 'product_photos_qty'])
dataframe
```

```python
train, test = train_test_split(dataframe, test_size = 0.2)
train.dtypes
```

```python
# Splitting my dataset into a train and test group

x_train = train.drop(columns = ['price'])
y_train = train['price']

x_test = test.drop(columns = ['price'])
y_test = test['price']
```

```python
x_train
```

```python
# Building my Linear Regression Model

from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X = x_train, y =  y_train)

pred = model.predict(x_test)

rmse = np.mean((np.sqrt((pred - y_test)**2)))

print(rmse)
```
